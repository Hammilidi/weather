```python
import requests
from datetime import datetime

# Remplacez 'YOUR_API_KEY' par votre clé API OpenWeather
api_key = '37e333ac92209e27137831ba501a0c03'

# Liste de villes au Maroc pour lesquelles vous souhaitez obtenir des données météorologiques
cities = ['Casablanca', 'Rabat', 'Marrakech', 'Fes', 'Tangier']

# Code de pays pour le Maroc
country_code = 'MA'

# Nombre de jours de prévision souhaité
num_days = 5

# Parcourez la liste de villes et effectuez des requêtes pour chaque ville
for city in cities:
    # URL de l'endpoint de l'API pour les prévisions météorologiques
    url = f'http://api.openweathermap.org/data/2.5/forecast?q={city},{country_code}&appid={api_key}'

    # Effectuez la requête HTTP pour obtenir les données de prévision
    response = requests.get(url)

    # Vérifiez la réponse
    if response.status_code == 200:
        data = response.json()
        # Les données de prévision sont stockées dans 'data'
        print(f'Prévisions météorologiques pour {city} pour les prochains {num_days} jours:')
        for forecast in data['list'][:num_days]:
            # Date de la prévision
            timestamp = forecast['dt']
            date = datetime.utcfromtimestamp(timestamp).strftime('%Y-%m-%d %H:%M:%S')
            print(f'Date: {date}')
            # Les autres informations météorologiques pour cette date
            print(f'Données météorologiques: {forecast}')
            print('\n')
    else:
        print(f'Erreur pour {city}: {response.status_code}')
```

## National Centers for Environment Information

```python
import requests
import os
from tqdm import tqdm
from io import BytesIO
import gzip

# Définissez les valeurs des variables directement
USAF = 12345  # Remplacez par le code USAF de votre choix
START_YEAR = 2022  # Année de début
END_YEAR = 2023  # Année de fin

# Calculez le nombre total de fichiers à télécharger
TOTAL_FILES = END_YEAR - START_YEAR + 1

BASE_URL = "https://www1.ncdc.noaa.gov/pub/data/noaa"

# Vérifiez si le répertoire "downloads" existe, sinon, créez-le
if not os.path.exists("./raw_data"):
    os.makedirs("./raw_data")

# Initialisez les compteurs de fichiers téléchargés et sautés
downloaded = 0
skipped = 0

# Téléchargez les fichiers
print("Downloading files...")
for i in tqdm(range(START_YEAR, END_YEAR + 1)):
    file_name = f"{USAF}-99999-{i}.gz"
    file_link = f"{BASE_URL}/{i}/{file_name}"

    # Si le fichier existe déjà, passez au suivant
    if os.path.exists(f"./downloads/{file_name}"):
        downloaded += 1
        continue

    try:
        response = requests.get(file_link)
        content = BytesIO(response.content)
    except:
        skipped += 1
        continue

    with gzip.GzipFile(fileobj=content) as f:
        extracted_data = f.read()

    # Écrivez le fichier téléchargé dans le répertoire "downloads"
    with open(f"./downloads/{file_name[:-3]}", "wb") as f:
        f.write(extracted_data)

    downloaded += 1

print(f"Successfully downloaded {downloaded} files out of {TOTAL_FILES}...")
print(f"Skipped {skipped} files out of {TOTAL_FILES}...")
```

```python
import requests
from bs4 import BeautifulSoup
import os

BASE_URL = "https://www1.ncdc.noaa.gov/pub/data/noaa"

# Créez un répertoire pour stocker les fichiers téléchargés
download_directory = "noaa_data"
os.makedirs(download_directory, exist_ok=True)

# Parcourez les années de 2000 à 2023
for year in range(2000, 2024):
    # Créez l'URL de l'année
    year_url = f"{BASE_URL}/{year}/"

    # Obtenez le contenu HTML de la page
    response = requests.get(year_url)
    if response.status_code == 200:
        # Analysez le contenu HTML avec BeautifulSoup
        soup = BeautifulSoup(response.text, "html.parser")

        # Parcourez les liens sur la page
        for link in soup.find_all("a"):
            file_url = link.get("href")
            if file_url.endswith(".gz"):
                # Téléchargez le fichier
                file_name = file_url.split("/")[-1]
                file_path = os.path.join(download_directory, file_name)
                file_url = f"{year_url}{file_url}"
                response = requests.get(file_url)
                if response.status_code == 200:
                    with open(file_path, "wb") as file:
                        file.write(response.content)
                    print(f"Téléchargé : {file_name}")
    else:
        print(f"Impossible d'accéder à {year_url}")

print("Téléchargement terminé.")
```
