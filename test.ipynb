```python
import requests

# Clé API OpenWeatherMap
api_key = '37e333ac92209e27137831ba501a0c03'

# Ville et pays
city = "Casablanca"
country = "MA"

# Endpoint de l'API d'agriculture
agro_api_endpoint = "http://api.openweathermap.org/data/2.5/agro/forecast"

# Paramètres de requête
params = {
    "q": f"{city},{country}",
    "appid": api_key,
}

# Effectuer la requête HTTP pour obtenir les données d'agriculture
response = requests.get(agro_api_endpoint, params=params)

if response.status_code == 200:
    agro_data = response.json()
    # Vous pouvez accéder aux informations spécifiques de l'API d'agriculture ici
    print("Données d'agriculture :")
    print(agro_data)
else:
    print(f"Erreur : {response.status_code}")
```
